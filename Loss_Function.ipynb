```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plotting MSE loss function where x_vals denotes predicted values and the actual value is set to 0

x_vals = np.arange(-10, 10, 0.01)
y_vals = np.square(x_vals)
plt.plot(x_vals, y_vals, "blue")
plt.title("MSE loss vs predictions")
plt.xlabel('Predictions')
plt.ylabel('Loss')
plt.show()
```

```python
# Plotting MAE loss function where x_vals denotes predicted values and the actual value is set to 0

x_vals = np.arange(-10, 10, 0.01)
y_vals = np.abs(x_vals)
plt.plot(x_vals, y_vals, "blue")
plt.title("MAE loss vs predcitions")
plt.xlabel('Predictions')
plt.ylabel('Loss')
plt.show()
```

```python
# Huber loss function
def huber_loss(y_pred, y, delta=1.0):
    huber_mse = 0.5*(y-y_pred)**2
    huber_mae = delta * (np.abs(y - y_pred) - 0.5 * delta)
    return np.where(np.abs(y - y_pred) <= delta, huber_mse, huber_mae)
    
# Plotting
x_vals = np.arange(-10, 10, 0.01)

delta_a = 0.1
huber_mse_a = 0.5*np.square(x_vals)
huber_mae_a = delta_a * (np.abs(x_vals) - 0.5 * delta_a)
y_vals_a = np.where(np.abs(x_vals) <= delta_a, huber_mse_a, huber_mae_a)

delta_b = 1
huber_mse_b = 0.5*np.square(x_vals)
huber_mae_b = delta_b * (np.abs(x_vals) - 0.5 * delta_b)
y_vals_b = np.where(np.abs(x_vals) <= delta_b, huber_mse_b, huber_mae_b)

delta_c = 10
huber_mse_c = 0.5*np.square(x_vals)
huber_mae_c = delta_c * (np.abs(x_vals) - 0.5 * delta_c)
y_vals_c = np.where(np.abs(x_vals) <= delta_c, huber_mse_c, huber_mae_c)

plt.plot(x_vals, y_vals_a, label = "0.1")
plt.plot(x_vals, y_vals_b, label = "1")
plt.plot(x_vals, y_vals_c, label = "10")
plt.title("Huber loss/ smooth MAE loss vs predicted values")
plt.xlabel('Predictions')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Plotting log Cosh loss function where x_vals denotes predicted values and the actual value is set to 0

x_vals = np.arange(-10, 10, 0.01)
x_vals_1 = np.cosh(x_vals)
y_vals = np.log(x_vals_1)
plt.plot(x_vals, y_vals, "blue")
plt.title("Log-cosh loss vs predicted values")
plt.xlabel("Predictions")
plt.ylabel("Loss")
plt.show()
```

```python
def compute_quantile_loss(y_true, y_pred, quantile):
    residual = y_true - y_pred
    return np.maximum(quantile * residual, (quantile - 1) * residual)

x_vals = np.arange(-10, 10, 0.01)
y_true = np.zeros(len(x_vals))

quantiles = [0.25, 0.5, 0.75]
quantile_losses = [
    compute_quantile_loss(y_true, x_vals, quantile)
    for quantile in quantiles
]

for quantile_loss in quantile_losses:
    plt.plot(x_vals, quantile_loss)

#plt.legend([str(int(q * 100)) + 'th percentile' for q in quantiles])
plt.legend([q for q in quantiles])
plt.xlabel('Predictions')
plt.ylabel('Loss')
plt.title('Quantile loss with predicted values (color quantiles)')
#plt.legend()
plt.show()
```

```python
# Cross entropy Loss 
def cross_entropy_loss(yHat, y):
    if y == 1:
      return -np.log(yHat)
    else:
      return -np.log(1 - yHat)


x_vals = np.arange(0, 1, 0.0001)
cost_1 = cross_entropy_loss(x_vals, 1)
cost_0 = cross_entropy_loss(x_vals, 0)

plt.plot(x_vals, cost_1, label='y=1')
#plt.plot(x_vals, cost_0, label='y=0')
plt.xlabel('Predictions')
plt.ylabel('Loss')
plt.title('Cross entropy loss with predicted values')
#plt.legend(loc='best')
#plt.tight_layout()
plt.show()
```
